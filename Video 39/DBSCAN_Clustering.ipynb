# <center style='color:orchid'> `DBSCAN` clustering (using `Scikit-Learn`) </center>

## 1. Import required libraries

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn import datasets
from sklearn.cluster import DBSCAN
from sklearn import metrics
```

## 2. Create data

```python
X, y = datasets.make_moons(n_samples=500, shuffle=False, noise=0.07, random_state=42)
X.shape, y.shape
```

## 3. Visualize data

```python
plt.figure(figsize=(10, 6), facecolor='honeydew')
sc = plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='brg')
plt.xlabel('X axis', fontsize=23)
plt.ylabel('Y axis', fontsize=23)
plt.title('Data partition', fontsize=27)
plt.legend(*sc.legend_elements(), title='Partition', fontsize=18, title_fontsize=21, bbox_to_anchor=(1.205, 0.287));
```

# <span style='color:magenta'> Case 1 </span>

## 4. Perform DBSCAN clustering (case 1)

```python
dbscan1 = DBSCAN(eps=0.2, min_samples=5)
dbscan1.fit(X)
labels1 = dbscan1.labels_
```

```python
np.unique(labels1)
```

```python
np.unique(y, return_counts=True)
```

```python
np.unique(labels1, return_counts=True)
```

```python
(y == labels1).all()
```

```python
metrics.accuracy_score(y, labels1)
```

## 5. Visualize DBSCAN clustering results (case 1)

```python
plt.figure(figsize=(10, 6), facecolor='lavenderblush')
sc1 = plt.scatter(X[:, 0], X[:, 1], c=labels1, s=50, cmap='gnuplot')
plt.xlabel('X axis', fontsize=23)
plt.ylabel('Y axis', fontsize=23)
plt.title('DBSCAN clustering results (case 1)', fontsize=27)
plt.legend(*sc1.legend_elements(), title='Clusters', fontsize=18, title_fontsize=21, bbox_to_anchor=(1.20, 0.287));
```

# <span style='color:crimson'> Case 2 </span>

## 6. Perform DBSCAN clustering (case 2)

```python
dbscan2 = DBSCAN(eps=0.15, min_samples=5)
dbscan2.fit(X)
labels2 = dbscan2.labels_
```
