# <center style='color:mediumpurple'> `OPTICS` clustering (using `Scikit-Learn`) </center>

## 1. Import required libraries

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn import datasets
from sklearn.cluster import OPTICS
from sklearn import metrics
```

## 2. Create data

```python
X, y = datasets.make_circles(n_samples=500, shuffle=False, factor=0.4, noise=0.03, random_state=42)
X.shape, y.shape
```

## 3. Visualize data

```python
plt.figure(figsize=(10, 6), facecolor='seashell')
sc = plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='cividis')
plt.xlabel('X axis', fontsize=23)
plt.ylabel('Y axis', fontsize=23)
plt.title('Data partition', fontsize=27)
plt.legend(*sc.legend_elements(), title='Partition', fontsize=18, title_fontsize=21, bbox_to_anchor=(1.202, 0.287));
```

# <span style='color:yellow'> Case 1 </span>

## 4. Perform OPTICS clustering (case 1)

```python
optics1 = OPTICS(max_eps=2, min_samples=24, xi=0.05)
optics1.fit(X)
labels1 = optics1.labels_
```

```python
np.unique(labels1)
```

```python
np.unique(y, return_counts=True)
```

```python
np.unique(labels1, return_counts=True)
```

```python
(y == labels1).all()
```

```python
metrics.accuracy_score(y, labels1)
```

## 5. Visualize OPTICS clustering results (case 1)

```python
plt.figure(figsize=(10, 6), facecolor='whitesmoke')
sc1 = plt.scatter(X[:, 0], X[:, 1], c=labels1, s=50, cmap='plasma')
plt.xlabel('X axis', fontsize=23)
plt.ylabel('Y axis', fontsize=23)
plt.title('OPTICS clustering results (case 1)', fontsize=27)
plt.legend(*sc1.legend_elements(), title='Clusters', fontsize=18, title_fontsize=21, bbox_to_anchor=(1.198, 0.287));
```

```python
plt.figure(figsize=(7, 3), facecolor='azure')
plt.plot(optics1.labels_[optics1.ordering_], color='navy', linewidth=3)
plt.xlabel('Samples', fontsize=16)
plt.ylabel('Labels', fontsize=16)
plt.yticks([0, 1], fontsize=13)
plt.xticks([0, 250, 500], fontsize=13)
plt.title('OPTICS clustering label ordering (case 1)', fontsize=19);
```

# <span style='color:tan'> Case 2 </span>

## 6. Perform OPTICS clustering (case 2)

```python
optics2 = OPTICS(max_eps=2, min_samples=24, xi=0.034)
optics2.fit(X)
labels2 = optics2.labels_
```

```python
np.unique(labels2)
```

```python
np.unique(y, return_counts=True)
```

```python
np.unique(labels2, return_counts=True)
```

```python
(y == labels2).all()
```

```python
metrics.accuracy_score(y, labels2)
```

## 7. Visualize OPTICS clustering results (case 2)

```python
plt.figure(figsize=(10, 6), facecolor='snow')
sc2 = plt.scatter(X[:, 0], X[:, 1], c=labels2, s=50, cmap='cool')
plt.xlabel('X axis', fontsize=23)
plt.ylabel('Y axis', fontsize=23)
plt.title('OPTICS clustering results (case 2)', fontsize=27)
plt.legend(*sc2.legend_elements(), title='Clusters', fontsize=18, title_fontsize=21, bbox_to_anchor=(1.1997, 0.366));
```

```python
plt.figure(figsize=(7, 3), facecolor='lightgoldenrodyellow')
plt.plot(optics2.labels_[optics2.ordering_], color='deeppink', linewidth=3)
plt.xlabel('Samples', fontsize=16)
plt.ylabel('Labels', fontsize=16)
plt.yticks([-1, 0, 1], fontsize=13)
plt.xticks([0, 50, 100, 150, 200, 250, 300, 350, 400, 450, 500], fontsize=13)
plt.title('OPTICS clustering label ordering (case 2)', fontsize=19);
```
