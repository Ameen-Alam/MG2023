## 4. Define dataloaders

```python
training_dataloader = DataLoader(training_dataset, batch_size=32, shuffle=True, drop_last=True)
test_dataloader = DataLoader(test_dataset, batch_size=32, shuffle=False, drop_last=True)
```

## 5. Define the network architecture

```python
class VAE(nn.Module):
    def __init__(self):
        super(VAE, self).__init__()
        
        self.encoder = nn.Sequential()
        self.encoder.add_module('linear1', nn.Linear(in_features=16*16, out_features=128))
        self.encoder.add_module('relu1', nn.ReLU(inplace=True))
        self.encoder.add_module('linear2', nn.Linear(in_features=128, out_features=64))
        self.encoder.add_module('relu2', nn.ReLU(inplace=True))
        self.encoder.add_module('linear3', nn.Linear(in_features=64, out_features=64))
        
        self._mu = nn.Linear(in_features=64, out_features=30)
        self._logvar = nn.Linear(in_features=64, out_features=30)
        
        self.decoder = nn.Sequential()
        self.decoder.add_module('linear3', nn.Linear(in_features=30, out_features=64))
        self.decoder.add_module('relu3', nn.ReLU(inplace=True))
        self.decoder.add_module('linear2', nn.Linear(in_features=64, out_features=128))
        self.decoder.add_module('relu2', nn.ReLU(inplace=True))
        self.decoder.add_module('linear1', nn.Linear(in_features=128, out_features=16*16))
        self.decoder.add_module('sigmoid1', nn.Sigmoid())
        
    def reparameterization(self, mu, logvar):
        std = torch.exp(0.5 * logvar)
        eps = torch.rand_like(std)
        sampling = mu + (eps * std)
        return sampling
        
    def forward(self, x):
        x = self.encoder(x.view(x.shape[0], -1))
        
        mu = self._mu(x)
        logvar = self._logvar(x)
        
        x = self.reparameterization(mu, logvar)
        
        return self.decoder(x), mu, logvar
```

In `VAEs`, `Encoder` block outputs two vectors: `mu` and `logvar`. These two vectors are two distributions basically. From these two distributions, latent space is sampled as a __Gaussian distribution__.

## 6. Visualize the model architecture

```python
model = VAE().to(device)
summary(model, input_size=(32, 1, 16, 16))
```

## 7. Define the training batch

```python
def training_batch(data, model, optimizer):
    model.train()
    data = data.to(device)
    recon, mu, logvar = model(data)
    loss = loss_function(recon, data, mu, logvar)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    return loss
```

## 8. Define the test batch

```python
with torch.inference_mode():
    def test_batch(data, model):
        model.eval()
        data = data.to(device)
        recon, mu, logvar = model(data)
        loss = loss_function(recon, data, mu, logvar)
        return loss
```

## 9. Set up model (`VAE`), criterion (`loss function`) and optimizer

```python
def loss_function(recon, x, mu, logvar):
    RECON = F.mse_loss(recon, x.view(x.shape[0], -1), reduction='sum')
    KLD = -0.5 * torch.sum(1 + logvar - mu.pow(2) - logvar.exp())
    return RECON + KLD
```

```python
vae = VAE().to(device)
optimizer = optim.Adam(vae.parameters(), lr=0.001)
```

## 10. Train the model

```python
np.random.seed(42)
torch.manual_seed(42)
torch.cuda.manual_seed(42)

n_epochs = 100

training_loss, test_loss = [], []

for epoch in range(n_epochs):
    training_losses, test_losses = [], []
    
    for data, _ in training_dataloader:
        trng_batch_loss = training_batch(data, vae, optimizer)
        training_losses.append(trng_batch_loss.item())
    training_per_epoch_loss = np.array(training_losses).mean()
    
    for data, _ in test_dataloader:
        tst_batch_loss = test_batch(data, vae)
        test_losses.append(tst_batch_loss.item())
    test_per_epoch_loss = np.array(test_losses).mean()
    
    training_loss.append(training_per_epoch_loss)
    test_loss.append(test_per_epoch_loss)
    
    if (epoch+1) % 10==0:
        print(f'Epoch: {epoch+1}/{n_epochs}\t| Training loss: {training_per_epoch_loss:.4f} |   ', end='')
        print(f'Test loss: {test_per_epoch_loss:.4f}')
```

I trained the `VAE` model for `100 epochs` only. However, you can obviously train it for a longer period of time and observe if the performance gets better or not! 

## 11. Plot model loss

```python
plt.figure(figsize=(8, 5))

plt.plot(training_loss, 'r-', linewidth=2, label='Training loss')
plt.plot(test_loss, 'b--', linewidth=2, label='Test loss')
plt.title('Loss curve', fontsize=23)
plt.xlabel('No. of epochs', fontsize=17)
plt.ylabel('Losses', fontsize=17)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(fontsize=14);
```

```python
ORG, REC = [], []

for a in range(8):
    idx = np.random.randint(len(test_dataset))
    ORG.append(test_dataset[idx][0])
    REC.append(vae.cpu()(ORG[a]))
    
fig, ax = plt.subplots(nrows=2, ncols=8, figsize=(5, 1))

for i in range(8):
    ax[0, i].imshow(ORG[i].squeeze(), cmap='Greys_r')
    ax[0, i].set_xticks([])
    ax[0, i].set_yticks([])
    ax[1, i].imshow(REC[i][0].reshape(16, 16).squeeze().cpu().detach().numpy(), cmap='Greys_r')
    ax[1, i].set_xticks([])
    ax[1, i].set_yticks([])

ax[0, 0].set_ylabel('Original', fontsize=12, rotation=0, labelpad=24, fontweight='bold')
ax[1, 0].set_ylabel('Reconstruction', fontsize=12, rotation=0, labelpad=48, fontweight='bold');
```
