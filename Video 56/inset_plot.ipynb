# <center style='color:coral'>Inset plotting using Matplotlib</center>

## 1. Import required libraries

```python
import matplotlib.pyplot as plt
import numpy as np
```

## 2. Create data

```python
arr = np.arange(100)
sin = np.sin(2*np.pi*(arr/100))
```

## 3. Plot `Sine Wave`

```python
plt.figure(figsize=(7, 4))
plt.plot(arr, sin, color='r', linewidth=3)
plt.xlabel('Time', fontsize=15, fontweight='bold')
plt.ylabel('Amplitude', fontsize=15, fontweight='bold')
plt.title('Sine Wave', fontsize=20, fontweight='bold');
```

## 4. Inset plotting

```python
# zoomed-in view

fig, out = plt.subplots(figsize=(10, 4))
out.plot(arr, sin, color='r', linewidth=3)
out.set_xlabel('Time', fontsize=15, fontweight='bold')
out.set_ylabel('Amplitude', fontsize=15, fontweight='bold')
out.set_title('Sine Wave', fontsize=20, fontweight='bold')
out.set_xlim(0, 50)
out.set_ylim(0, 1)

# original view

inset = fig.add_axes([0.35, 0.3, 0.35, 0.35])  # [left, bottom, width, height]
inset.plot(arr, sin, color='g', linewidth=2)

# zoomed-in region

rect = plt.Rectangle((0, 0), 50, 1, linewidth=2, edgecolor='b', facecolor='None')
inset.add_patch(rect);
```
